# The Two Body Problem

The general two-body problem consists of solving for the motion of two masses that move under the influence of each other’s gravity. In orbital mechanics, typically one of the masses is considered to have a negligible mass, causing it to have no effect on the other mass. The motion of the negligible mass is of interest and what is solved for. In reality, the smaller mass does have an effect on the larger mass, but the effect is so minuscule that it is, for all practical purposes, zero.

## Derivation of equations of motion
If we apply the Newton's second law equations to each of the particles, using the diagram to assist, we will get the following equations of motion for each of the masses:

 $$m_{1}\ddot{\overrightarrow{R_1}}=\frac{Gm_{1}m_{2}}{R^{3}_{12}}\overrightarrow{R_{12}}$$
 $$m_{2}\ddot{\overrightarrow{R_2}}=-\frac{Gm_{1}m_{2}}{R^{3}_{12}}\overrightarrow{R_{12}}$$

 , where $G$ is the **universal gravitational constant**. This is a physical constant which determines the intensity of the gravitational force between two body masses. Its value is $$G=6.678e^{-11} Nm^2/kg$$.

 
The **$R$** vectors give the inertial, second time derivative for each of the masses. If we cancel the masses on each side of the equations and change the **$R_{12}$** vector to just **$r$**, we get the following:

 $$\ddot{\overrightarrow{R_1}}=-\frac{Gm_{2}}{r^{3}}\overrightarrow{r}$$
 $$\ddot{\overrightarrow{R_2}}=-\frac{Gm_{1}}{r^{3}}\overrightarrow{r}$$
Since
 $$\ddot{\overrightarrow{r}}=\ddot{\overrightarrow{R_1}}-\ddot{\overrightarrow{R_2}}$$
follows that
 $$\ddot{\overrightarrow{r}}=-\frac{Gm_{1}}{r^{3}}\overrightarrow{r}-\frac{Gm_{2}}{r^{3}}\overrightarrow{r} = -\frac{(Gm_{1} + Gm_{2})}{r^{3}} \overrightarrow{r}$$
 
We are typically trying to solve for the trajectory of a satellite around a much larger mass, so we can make the assumption that the mass of the satellite is negligible compared to the major body ($m_2 << m_1$). Also, we can define the **gravitational parameter  $\mu$ as $\mu = G*m$**.

By doing this, the last equation can be written as:

 $$\ddot{\overrightarrow{\mathbf{r}}}=-\frac{\mu }{r^{3}}\overrightarrow{\mathbf{r}} $$

## Constants of the motion
The two-body problem has a set of constants of motions that can be derived from the above equations. Here below the results:

### Orbital angular momentum
It is defined as: 
$$\overrightarrow{h}=\overrightarrow{r}\times \overrightarrow{v}$$
and is constant over time. This implies the motion of the two masses always belong to the same plane

<center><img src="Images/angularMomentum.png" alt="Drawing" style="width: 550px;"/>






### Orbital energy per unit mass
It is defined as:
$$\varepsilon= \frac{v^{2}}{2}-\frac{\mu}{r}$$
and is constant as well. Once defined the state vector (position and velocity) of $m_{2}$ respect to $m_{1}$ at initial time, we also define the mechanical energy of the orbit itself. Note that $\varepsilon$ can be negative, null or positive.

### Eccentricity vector
It is defined as:
$$\overrightarrow{e}=\frac{\overrightarrow{v}\times \overrightarrow{h}}{\mu}-\frac{\overrightarrow{r}}{r}$$
it lies in the orbital plane and is fixed respect to it. The line defined by the vector $\overrightarrow{e}$ is commonly called the **apse line**.

## The Orbit Equation

We have established that Keplerian orbital motion is planar and the mechanical energy is constant. The orbital plane can be specified by the angular momentum vector $\overrightarrow{h}$. We also found a vector that is fixed in the orbital plane, that is the eccentricity vector $\overrightarrow{e}$.

We now define the **true anomaly $\theta$** as the angle between the position vector and the eccentricity vector:

 $$cos(\theta)=\frac{\overrightarrow{r}\cdot\overrightarrow{e}}{re}$$

<center><img src="Images/trueAnomaly.png" alt="Drawing" style="width: 380px;"/><center>

&nbsp;  
After some derivations of the above equation, we finally get the **orbit equation**, defining the path of $m_{2}$ around $m_{1}$. 

$$r=\frac{p}{1 + e\cos\theta}=\frac{h^{2}}{\mu}\frac{1}{1 + e\cos\theta}$$


In this equation **$h$**, **$e$** and **$\mu$** are all constant. Since $e$ is the magnitude of $\overrightarrow{e}$, it is strictly positive. $p$ is called **semilatus rectum**.

The **flight path angle** $\gamma$ is the angle between the velocity vector and the local horizon (measured positive in the positive r-direction)

$$\tan\gamma=\frac{e\sin\theta}{1+e\cos\theta}$$


## The Vis-viva Equation

The Vis-viva equation is a fundamental equation used in orbital mechanics which relates the velocity of an object in an orbit around a planet, to its distance from the center of the planet and the semi-major axis of its orbit. It is the direct result of the principle of conservation of mechanical energy along the orbit.

 $$v^{2}=\mu(\frac{2}{r}-\frac{1}{a})$$

This equation has many implications, like the following:

**Orbital velocity**: To determine the orbital velocity of an object in a circular orbit of radius $r$, just substitute $a = r$ in the above equation and you get:

 $$v=\sqrt{\frac{\mu}{r}}$$
**Escape velocity**: To determine escape velocity the object will have to reach infinity ($a = \infty$) starting from a distance $r$ from the attracting body:

 $$v_{esc}=\sqrt{\frac{2\mu}{r}}$$

Replacing $r$ with the Earth radius, the escape velocity is about 11.18 km/sec.

Vis-viva also makes easy to calculate the Delta-v required for orbit maneuvers.

# Shape of a Keplerian orbit

The resulting path (position over time) resulting from the orbit equation is called **Keplerian orbit**. 
The orbit equation describes conic sections, meaning that all orbits are one of four types. The particular type of orbit is determined by the magnitude of the eccentricity:

1- Circles: **$e=0$**

2- Ellipses: **$0<e<1$**

3- Parabolas: **$e=1$**

4- Hyperbolas **$e>1$**

<center><img src="Images/TypesOfConicSections.png" alt="Drawing" style="width: 600px;"/><center>





## Circular Orbits
Setting **$e=0$** in the orbit equation we get:

 $$r=\frac{h^{2}}{\mu}$$
 
The radius is constant, so the orbit is a circle. The velocity is also constant:

 $$v=\sqrt{\frac{\mu}{r}}$$
 
The time required to complete one orbit is known as the **orbital period**. Because the speed of a circular orbit is constant, the period is simply the circumference divided by the velocity:

 $$T=2\pi\sqrt{\frac{r^3}{\mu}}$$
 
The specific energy of a circular orbit is:

 $$\varepsilon= -\frac{\mu}{2r}$$

It is always negative, meaning that the motion is resembling a closed pattern (circular in this case)

## Elliptical Orbits
If the eccentricity is between 0 and 1, then the radius of the orbit varies with the true anomaly and the orbit has an elliptical shape. The minimum value of $r$ occurs at periapsis where $\theta=0$ and the maximum value of $r$ is at apoapsis where $\theta=\pi$.
We also define the **semimajor axis $a$** as the distance between the center of the ellipse and its furthest edge
<center><img src="Images/elliptical.png" alt="Drawing" style="width: 600px;"/><center>


Here below are shown the most important equations relating the different elements of the ellipse
$$a=\frac{h^2}{\mu}\frac{1}{1-{e^{2}}}$$

 $$r_{p}=\frac{h^2}{\mu}\frac{1}{1+e}=a(1-e)$$

$$r_{a}=\frac{h^2}{\mu}\frac{1}{1-e}=a(1+e)$$

 $$\frac{r_{p}}{r_{}a}=\frac{1-e}{1+e}$$

 $$\frac{v^{2}}{2}-\frac{\mu}{r}=-\frac{\mu}{2a}$$

$$e=\frac{r_{a}-r_{p}}{r_{a}+r_{p}}$$

Also, the period $T$ and the semimajor axis $a$ are related related each other, and they don't depend on the eccentricity $e$:

 $$T=2\pi\sqrt{\frac{a^3}{\mu}}$$

The specific energy of a parabolic orbit is:

 $$\varepsilon= -\frac{\mu}{2a}$$


## Parabolic Orbits
Setting **$e=1$** in the orbit equation we get:

 $$r=\frac{h^{2}}{\mu}\frac{1}{1+\cos\theta}$$

As the true anomaly $\theta$ approaches 180 deg, the denominator approaches zero, so that $r$ tends towards infinity. According with the equation

 $$\varepsilon= -\frac{1}{2}\frac{\mu^{2}}{h^{2}}(1-e^{2})$$

the specific energy of a trajectory for which $e =1$ is zero, so that for a parabolic trajectory the conservation of energy become

 $$\frac{v^{2}}{2}-\frac{\mu}{r} = 0$$

...and the velocity for each position $r$ is

 $$v=\sqrt{\frac{2\mu}{r}}$$
 
If the body $m_{2}$ is launched on a parabolic trajectory, it will coast to infinity, arriving there with zero velocity relative to $m_{1}$. Parabolic paths are therefore called **escape orbits**.

## Hyperbolic Orbits
For **$e>1$** in the orbit equation we get an hyperbola:

 $$r=\frac{h^{2}}{\mu}\frac{1}{1+e\cos\theta}$$

The denominator goes to zero when $1+ecos(\theta)=0$. The relevant true anomaly $\theta_{\infty}$ is called **true anomaly of the asymptote**. A characteristic parameter is the angle between the asymptotes and the $x$ axis:

 $$\beta=cos^{-1}(\frac{1}{e})$$

This means that once the eccentricity is defined, the shape of the hyperbolic orbit is also defined.

<center><img src="Images/hyperbolic.png" alt="Drawing" style="width: 600px;"/><center>

The angle $\delta$ between asymptotes is called **turn angle**. This is the angle through which the velocity vector of the orbiting body is rotated as it rounds the attracting body at F and heads back towards infinity.

The specific energy of a parabolic orbit is positive and equal to:

 $$\varepsilon= \frac{\mu}{2a}$$

The conservation of energy for a hyperbolic trajectory is

 $$\frac{v^{2}}{2}-\frac{\mu}{r}= \frac{\mu}{2a}$$

Starting from the above equation, let $v_{\infty}$ denote the speed at which a body on a hyperbolic path arrives at infinity ($r=\infty$):

 $$v_{\infty}=\sqrt{\frac{2\mu}{a}}$$

$v_{\infty}$ is called **hyperbolic eccess velocity**. It represents the amount of extra specific energy respect to energy needed to reach the escape velocity. 

The hyperbolic eccess velocity can also be related to the escape velocity using the following relationship:

 $$v_{\infty}=\sqrt{v^{2}-v_{esc}^{2}}$$

A relatively small extra velocity above that needed to accelerate to the escape speed can result in a relatively large speed at infinity. For example, a satellite on a circular orbit at 300 km altitude goes at 7.72 km/sec. The escape velocity for Earth is 11.18 km/sec, so the needed extra velocity to get into an escape trajectory is 11.18 - 7.72 = 3.46 km/sec. However, applying an extra velocity of 4 km/sec (0.54 km/sec higher than the previous), we get hyperbolic eccess velocity of 3.52 km/sec




The square of $v_{\infty}$ is denoted **C3**, and is known as the characteristic energy:
 $$C3 = v_{\infty}^{2}$$
