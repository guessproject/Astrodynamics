# Lunar Orbits

## Moon Data

### Gravity Field

The Moon's gravity field, like Earth's, can be described using a mathematical representation known as **gravity field harmonics** (or spherical harmonics). This method models the gravitational potential as a series of terms, each representing a different-sized "lump" or feature in the field. While the complete description is made in Chapter 4 (Effects of Non-Spherical Earth), the mathematical formula describing the geo-potential is reported here for quick reference:


$$U(r, \phi, \lambda) = \frac{GM}{r} \left[ 1 + \sum_{l=2}^{\infty} \sum_{m=0}^{l} \left( \frac{R}{r} \right)^l P_{lm}(\sin\phi) (\bar{C}_{lm}\cos(m\lambda) + \bar{S}_{lm}\sin(m\lambda)) \right]$$



The key characteristics of the Moon's gravity field are the following:

1.  **Weakness**: the Moon's gravity is about **16.6%** (or one-sixth) of Earth's. This is because the Moon has a much smaller mass and is less dense than Earth
2.  **Lack of Atmosphere and Water**: unlike Earth, the Moon lacks significant fluid bodies (oceans and a thick atmosphere) that can redistribute mass and smooth out gravitational irregularities. This allows its uneven mass distribution to have a more direct and pronounced effect on its gravity field
3.  **Mass Concentrations**: this is the most defining feature of the Moon's gravity. These are regions of higher-than-average gravity that are concentrated in the center of the large, dark lunar seas
4.  **Influence on Spacecraft Orbits**: the lumpy gravity field makes it difficult to maintain a stable, low orbit around the Moon. Satellites in low lunar orbit can be pulled toward the mass concentrations, causing their orbits to become unstable and eventually decay. This is why missions like NASA's Lunar Reconnaissance Orbiter (LRO) require frequent orbital adjustments to avoid crashing into the Moon
5.  **Lack of an Equatorial Bulge**: Earth's rotation causes an equatorial bulge, which is the dominant perturbation to a satellite's orbit. The Moon's rotation is extremely slow (once every 27.3 days), so it does not have a similar bulge. The primary perturbation is from the mascons

### Gravity Models

The first lunar gravity field models (up to 13x13) were calculated from data obtained in 1966 and 1967 by the Soviet Luna 10 mission and the first four U.S. Lunar Orbiter missions. The U.S. Apollo program allowed for further study of the lunar gravity field with Apollo 15 and 16, leading to the 1993 Lun60d gravity model of degree and order 60. Further elaboration of available data broight the definition of more sophisticated models, such as

* **LP150Q** - The LP150Q gravity model was one of the primary models used for navigating spacecraft in lunar orbit during the early phases of missions like the Lunar Reconnaissance Orbiter (LRO). The "150" in LP150Q indicates that the model uses terms up to degree and order 150. The model's coefficients were derived by fitting the model to a vast amount of tracking data from lunar orbiting spacecraft.

* **GRAIL** - LP150Q was a significant improvement over earlier models and was able to meet the navigation accuracy requirements for the nominal phase of the LRO mission. However, its limitations became apparent during the extended mission phase, particularly in elliptical orbits. This led to the development of more advanced gravity models, such as those derived from the **GRAIL (Gravity Recovery and Interior Laboratory)** mission, which provided far more accurate data and enabled the creation of even higher-resolution models (e.g., up to degree 1200), offering a much better understanding of the lunar gravity field. A description of this gravity model is available at https://pgda.gsfc.nasa.gov/products/50 .


The derivation of gravity field models from real-world missions includes a series of complicating factors, the biggest of these is the fact that none of the models developed to this point include tracking data over the far side of the Moon, since direct communication with Earth there is impossible. Despite this, the lunar gravity field to low degree and order is considered stable and reliable, and can be used to plan lunar missions. ### Shape 

The general shape of the Moon is very nearly a perfect sphere, excluding local topography variations, which magnitudes are much larger than the overall flattening of the lunar poles (about 2 km) or any ellipticity of the lunar equator.

The IAU/IAG 2000 Report recommends using the same radius value (1737.4 km) for both the lunar equator and the lunar pole and to express topography data relative to a sphere with the same radius (reference sphere). ### Orbit

The Moon's orbit around the Earth is affected by several disturbances:

- The gravitational attraction of the Sun
- Perturbations (on the radial distance of the Moon) caused by Jupiter and Venus
- Earth’s $J_2$
- Solar Radiation Pressure

As a result, the orbit of the Moon is very complex and cannot be described using the simple “precessing, inclined ellipse” model so often used to describe the orbits of low-altitude Earth satellites.

In simple terms, the lunar orbit is slightly elliptical (mean eccentricity = 0.05490) with a mean semi-major axis of about 384,400 km. Its inclination is nearly constant with respect to the ecliptic (mean inclination = 5.15 deg), but its inclination with respect to the Earth’s equator varies from approximately 18-29 deg. All three of these orbital elements ($a$, $e$, and $i$) are subject to periodic variations due to solar perturbations.
The remaining orbital elements describing the orientation of the lunar orbit with respect to a particular coordinate system (i.e. longitude of the ascending node and argument of periapsis) are both affected by  secular and periodic changes.

### Low Lunar Orbits

A generic low-altitude lunar orbit (often below 100 km) is highly unstable and subject to rapid, significant changes in its Keplerian elements due to the Moon's highly irregular gravity field. The primary cause of these changes are the mass concentrations (**mascons**) beneath the lunar seas.

When in LLO, the Keplerian elements experience sudden and large changes, making orbital prediction and maintenance challenging. Here below the main trends are discussed, using as example the LRO (Lunar Reconnaissaince Orbiter) data (available from the NAIF website https://naif.jpl.nasa.gov/naif/data_archived.html) for a time span of 3 months.

<b>1. Semi-Major Axis ($a$)</b>

The semi-major axis, which defines the size of the orbit, is not significantly affected, as the gravitational anomalies tend to be conservative (they do work over short periods but the net energy change over an orbit is often small). The small, secular (long-term) decrease in $a$ is primarily due to third-body perturbations, mainly from the Earth and the Sun, or from a very small amount of atmospheric drag at extremely low altitudes

$\quad$

<b>2. Argument of Periapsis ($\omega$) and Eccentricity ($e$)</b>

The argument of periapsis experiences rapid, non-periodic changes (drift and oscillation). This is primarily due to the localized pull of the mass concentrations. Unlike Earth, where the $J_2$ (equatorial bulge) effect causes a slow, predictable drift, the Moon's highly localized mass anomalies exert strong, varying forces that push the orbital ellipse around in an erratic fashion. As consequence, the eccentricity also changes, leading to the periselene altitude to decrease quickly. This instability is the primary danger, as the orbit will rapidly decay or crash into the Moon without frequent maneuvers

Both the inclination and the right ascension of the ascending node (RAAN) experience secular (long-term) and periodic changes due to third-body effects. The secular effect is due to the non-spherical Moon's gravity, similar to Earth but in less uniform way, while Earth and Sun's gravity periodically perturb the orbit (oscillations around the mean)

### Elliptical Lunar Frozen Orbits (ELFO)

The primary goal of this kind of orbit is long-term stability, minimizing the long-term changes in two key orbital elements: **argument of perigee $ω$** and **eccentricity $e$.** These orbits are achieved by balancing the effects of two major forces:

* **Lunar Non-Spherical Gravity**: the irregular mass distribution (mascons) causes the argument of periselene (ω) and eccentricity (e) to drift wildly, especially in low orbits
* **Third-Body Perturbations (Earth and Sun)**: the gravitational pull of the Earth and Sun creates periodic changes in all orbital elements, which can be the dominant stabilizing or destabilizing factor depending on the orbit's altitude and inclination

Elliptical frozen orbits are preferred over low circular orbits for specific missions because they offer unique benefits:

- **Long-Term Station-Keeping**: by minimizing the secular (long-term, non-periodic) drift of $e$ and $ω$, these orbits require significantly less fuel for station-keeping maneuvers compared to generic low lunar orbits
- The point of farthest approach (aposelene) is very high, allowing for **global communication coverage** and prolonged exposure to the deep space environment
- **Orbital Resonance**: stability is often found at specific altitudes where the satellite's orbital period is in a resonance with the Moon's rotation period or with the Earth's orbital period around the Moon


#### Analytical Solution

To allow for an analytical solution of this kind of orbits the following assumptions are made:
- Negligible mass of the satellite
- The presence of no other perturbations other than the third body (in this case Earth)
- A circular orbit of the Earth about the Moon

In addition, all short period terms are averaged out to form a mean element model. By applying these assumptions, a series of Lagrange planetary equations is developed that specifies the time variance of orbital elements due to the Earth third-body effect. It can be demonstrated that the conditions

$$  \frac{\mathrm{d} e}{\mathrm{d} t}=\frac{\mathrm{d} \omega}{\mathrm{d} t}=0 $$


are satisfied when 

#### $$ \begin{matrix}
\omega = 90\degree, 270\degree \\
e^{2}+\frac{5}{3}cos^{2}i=1\end{matrix} $$

For any inclination between $39.2\degree$ and $140.77\degree$ and when $\omega = 90\degree$ or $\omega = 270\degree$, there exist an eccentricity which can be used to set both $\omega$ and $e$ rates to zero, hence the frozen condition is satisfied.


Critical Stability Inclinations
Research has identified specific inclinations that offer enhanced stability for elliptical lunar orbits:

Inclinations near 27 deg and 62 deg (Prograde): These are known to have regions of stability where the J 2and J3 effects are manageable, offering a decent lifetime.

Near-Polar Inclinations (Retrograde): Highly inclined, retrograde orbits (inclination greater than 90 ∘, meaning the satellite orbits opposite to the Moon's rotation) tend to exhibit better long-term stability and are often preferred for very-long-duration missions.

The challenge remains that even the most stable frozen lunar orbit is only quasi-stable. The chaotic nature of the Moon's gravity means that while the orbit may be stable over months or years, it will eventually become unstable over decades, requiring eventual disposal or a dedicated maneuver.
