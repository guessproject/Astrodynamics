# Generalized Perturbations Analysis

So far, we have only discussed idealized orbits, where the solution of the Two Body problem is derived in closed form. In this case, the angular momentum **$\overrightarrow{h}$**, the eccentricity vector **$\overrightarrow{e}$**, and the orbital energy **$\varepsilon$** are constant over time and this is reflected to the invariability of the classical elements.

In reality, there are several sources of disturbances which make the problem way harder to solve:
- Non-spherical Earth
- Drag
- Third body gravity
- Solar Radiation pressure
- Tidal effects

In this case, the equation of motion can be written as

$$\ddot{\overrightarrow{r}}=-\frac{\mu }{r^{3}}\overrightarrow{r}+\frac{\overrightarrow{F}}{m}$$

, where $m$ is the mass of the spacecraft, and $\overrightarrow{F}$ the perturbing force.

The orbital elements affected by the perturbing force are defined as **osculating elements**; they are not constant anymore, but will vary over time. Each set of osculating elements would describe a Keplerian orbit if the perturbing force disappeared instantly. Therefore , each point on the trajectory has a corresponding set of osculating elements, and they include the effects of the perturbations.

In this chapter we'll discuss how to derive a generalized relationship between disturbing forces and the orbital elements, and which way they can vary over time.

 


## The VOP Equations

In the presence of perturbations angular momentum and energy of the satellite are not conserved, hence the orbit is no longer truly elliptic. However, deviations from the ellipse occur over long time-horizons and so we assume a quasi-stationary elliptic motion and include adjustments to
the ellipse in the form of orbit-averaged versions of $\dot{a}, \dot{i}, \dot{\Omega}, \dot{\omega} , \dot{e}$ and $\dot{M}$.

For many applications, it's convenient to express the rates of changes of the elements in terms of disturbing forces. This approach is called **Variation of Parameters (VOP)**, where we use the solution for the unperturbed system to represent the solution of the perturbed system. Both Lagrange and Gauss developed VOP methods to analyze perturbation effects. The latter approach is here considered, since it accounts for non-conservative disturbance forces.

In general, the perturbations are not pointed along the direction joining the two centers of mass. In this regard, we define a new coordinate system called **RTN** (Radial-Transverse-Normal), in such a way the perturbing force $\overrightarrow{F}$ can be expressed as follows:

$$\overrightarrow{F}=R\hat{e_R}+T\hat{e_T}+N\hat{e_N}$$

After many mathematical derivations, it can be demonstrated that the variations of the orbital elements respect to perturbations expressed in the RTN system are the following, known as the **Gaussian form** of the VOP equations:


#### Semimajor Axis

$$\frac{da}{dt}=\frac{2}{n\sqrt{1-e^2}}[e\sin(\theta)\textbf{R}+\frac{p}{r}\textbf{T}]$$


#### Eccentricity

$$\frac{de}{dt}=\frac{\sqrt{1-e^2}}{na}[\sin\theta\textbf{R}+(\cos\theta+\frac{e+\cos\theta}{1+e\cos\theta})\textbf{T}]$$

#### Inclination

 $$\frac{di}{dt}=\frac{r \cos u}{na^2\sqrt{1-e^2}}\textbf{N}$$

#### RAAN

 $$\frac{d\Omega}{dt}=\frac{r\sin u}{na^2\sqrt{1-e^2}\sin i}\textbf{N}$$

#### Argument of Perigee

$$ \frac{d\omega}{dt}=\frac{\sqrt{1-e^2}}{nae}[-\cos \theta\textbf{R}+\sin\theta(1+\frac{r}{p})\textbf{T}]-\frac{r \cos i \sin u}{h \sin i}\textbf{N}$$

#### Mean Motion

$$ \frac{dM}{dt}=\frac{1}{na^2e}[(p \cos \theta-2er)\textbf{R}-(p+r)\sin \theta\textbf{T}]$$

, where $\theta$ is the true anomaly, $u$ is the argument of latitude and $n$ is the mean motion. The above equations show how the orbit elements change their istantaneous value accordingly with the direction of the perturbing force. Here below few considerations:

- Perturbations along the radial and the tangential directions bring to a change of the semimajor axis (energy is not conserved)
- Perturbations along the normal of the orbit plane don't change the semimajor axis, the eccentricity and the mean motion
- Inclination and RAAN only change if we have a perturbation normal to the orbit plane
- Depending on the value of the argument of latitude, inclination and RAAN variations may be null even if a normal perturbation is applied

### Periodic and Secular Variations, Mean Elements

When examining the effects of perturbations on the orbital elements, we must characterize how they vary over time. Gravitational perturbations affect the trajectory at any point along its path. Therefore, the orbital elements $ \overline{el} = \begin{Bmatrix}
a & e & i & \Omega & \omega & \theta \\
\end{Bmatrix}$ vary periodically and significantly over each orbit. Therefore we average the effects over one orbital period $T$ and define the **mean orbital elements** as:

$${el}_i = \frac{1}{T}\int_{0}^{T} \theta_i \cdot dt = \frac{1}{2 \pi}\int_{0}^{2 \pi}\theta_i \cdot dM$$

, where $M$ is the mean anomaly. Integrating this way (over the fast variable true anomaly) all the orbital elements, we get the **short periodic** effects on each of the element. Despite those effects are significant, they often are opposite on the two halves of an orbit, such that overall they fully cancel out over one orbital period.

However oblateness perturbations constantly vary the argument of periapsis and the argument of perigee such that the line of apsides and the orbital plane constantly rotate. Those slow changes imply that over one orbit the perturbational effects at any $\theta$ are little different from those at $-\theta$. This imperfect cancellation of opposite contributions accumulates and leads to slow variations of the mean orbital elements, which have **long periodic** effects.

We can also remove these long-period variations by integration over both one revolution of the line of apsides and the line of nodes, which results in the so-called **secular** effects

$$\overline{el}_{i,sec} = \frac{1}{T}\int_{0}^{2\pi} \int_{0}^{2\pi}\overline{\theta}_i \cdot d\overline{\omega} \cdot d\overline{\Omega}dt$$

## Simplified General Perturbations Models and TLE

The **Simplified General Perturbations (SGP)** models are analytical techniques which describe the motion of a satellite subject to disturbing forces over some definite time interval.  They are called *general* because the original intent was to arrange a mathematical model valid for *all* the possible initial states; in reality, the lack of fidelity in the model used prevents this formulation to be effective in all passible orbit regimes.

### SGP4

The first developed model was just called SGP while the most used one, called **SGP4**, was deployed to increase the fidelity of the first model. It only include the largest perturbations: J2 to J5 zonal harmonics, simplified drag and third body and solar radiation pressure; those assumptions can severely limit the accuracy of SGP4 propagation. Because many of short-periodic perturbations are not taken into account, the maximum possible accuracy is limited.

SGP4 is implemented as a sequence of nested polynomials and trigonometric functions into which one simply plugs the end time, and the final result is calculated from it in one step: this was an "hard" requirement to save computational power in predicting future position of each object in the catalog. The mathematics behind the formulation looks quite messy, as highlighted from the official report: https://celestrak.org/NORAD/documentation/spacetrk.pdfhttps://adsabs.harvard.edu/full/1981CeMec..24..367H


It separates a satellite's motion into two components:

- **Secular Perturbations**: these are long-term, non-periodic changes in the orbital elements, primarily caused by the Earth's non-spherical gravitational field ($J2$ effect) and atmospheric drag. The SGP4 model uses the TLE's mean elements to compute these secular changes over time
- **Periodic Perturbations**: these are short-term, oscillatory variations in the orbit that occur within a single revolution. These are also caused by the Earth's oblateness. The SGP4 model uses a series of trigonometric functions to add these periodic variations back to the secularly-changing elements

The SGP4 algorithm works by taking the TLE parameters at the epoch and then analytically integrating the secular and periodic effects over the desired time span to produce a predicted position and velocity vector (state vector) for the satellite at any given time.

SGP models were specifically designed to work with TLEs (**Two-Line Element**). A TLE is a compact data format that contains the necessary orbital parameters for the SGP model to predict a satellite's future position; those parameters are derived from a batch Least Squares algorithm, which minimize the residuals of the 'true' ephemeris respect to the SGP force model.

The key parameters that are essential for the SGP4 model are:

- **Epoch**: the reference time of the provided state vector
- **Mean Motion**: the average orbital speed of the satellite, in revolutions per day. This is a central parameter for the SGP4's analytical equations
- **Drag Term ($B^*$ or $B$-star)**: This is an empirical parameter derived from observations that accounts for the effects of atmospheric drag. It's a key part of the SGP4 model, especially for LEO satellites where drag is a significant perturbation
- **Other Orbital Elements**: the TLE also contains mean values for the eccentricity, inclination, argument of perigee, and right ascension of the ascending node, which are used as the initial conditions for the SGP4 equations


The general approach for estimating SGP4 states and generating TLE’s is like any other trajectory fit process done in orbit determination: a set of “observations” are generated, an a priori state is determined, and the observations are processed as a set to minimize the residual between the propagated a priori state and the observations. This process is repeated until some convergence tolerance is satisfied.
The unique aspect in this context is that the propagation is done via the SGP4 algorithm: <u>the relationship between TLEs and SGP4 is symbiotic and exclusive</u>. 

TLEs are specifically crafted to be the starting point for the SGP4 model. The process of generating a TLE from tracking data involves "removing" the periodic variations in a very specific way, and the SGP4 model is designed to "add" those same periodic variations back. This means a TLE is not meant to be used with any other type of orbit propagator. Using TLE parameters in a high-fidelity numerical model, for instance, would lead to incorrect results.

### TLE Catalog Management

One of the most relevant limitations of the classical TLE format is the limited number of object identifiers available. The satellite catalog number is limited to 5 digits, which means that a maximum number of 99999 objects can be correclty stored in the catalog. Considering the current catalog increasing rate, there will be no more available IDs in the second half of 2026. To overcome this situation, two different approaches are emerging:

1) **The Alpha-5 TLE**

This approach extends the current numbering by replacing the first (integer) digit of the catalog ID with an alphabetical character (from A to Z). Each alphabetical character correspond to 2-digit number:

2) **The OMM Format**

The **Orbit Mean-Elements Message (OMM)**  is intended to allow replication of the TLE content in a CCSDS standard format, plus additional data (such as object's mass and size or covarinace matrix) and comments. The objectd IDs are replaced with the **9-digits format YYYY-NNNP**, where:

- YYYY is the year of launch
- NNN is the serial number of launch in year YYYY
- P identifies the part brought into space by the launch (more that one P is allowed)

### Semi-analytical Models: The SGP4-XP

Semi-analytical models extend the capability of pure analytical models by implementing numerical propagation which allows a more accurate description of the dynamic, non-conservative perturbing forces. The most affected part is the long-term effect of forces such as drag and SRP, which are not very well suited for an anlytical representation.

In 2020 the United States Space Force (USSF) released an implementation of the **SGP4-XP** (eXtended Perturbations) propagator, an advanced version of the existing SGP4 propagator designed to significantly improve the accuracy of its predecessor. The main changes in the model are:

- Improved lunar perturbation modeling
- New and more resonance modeling for different orbit regimes
- Solar radiation pressure (SRP) modeling for all orbit regimes
- The Geopotential model now includes the J5 zonal term
- The legacy WGS-72 terms are replaced with EGM-96 terms
- The static atmosphere model is replaced with the Jacchia-70 atmospheric density model

The new theory also accomodats the ballistic coefficient for atmospheric drag and the solar radiation pressure values in the TLE fields formerly used by the Bstart and mean motion double dot, respectively.

While the accuracy is improved by a factor of 10 (especially in LEO and MEO), the runtimes are just 50 to 100% longer than SGP4.
